```python
import pandas as pd
import pymongo
```

```python
client = pymongo.MongoClient("mongo:27017")
database = client['exercices']
collection = database['youtube']
```

```python
df_youtube = pd.read_csv("./data/USvideos.csv")
df_youtube
```

## Question 0

### Netoyer les données

```python
df_youtube["tags"] = df_youtube["tags"].apply(lambda x : x.split("|"))
```

### Importer les données

```python
collection.delete_many({})
collection.insert_many(df_youtube.to_dict("records"))
```

## Question 1  

```python
# Récupérer toutes les vidéos de la chaine apple
cursor = collection.find({"channel_title" : "Apple"})
#next(cursor)
for document in cursor :
    print('-----')
    print(document)
```

## Question 4

```python
cur = collection.find().sort([("views", -1)]).limit(15)
list(cur)
```

## Question 5

```python
cur = collection.aggregate([{"$group":{"_id":"$category_id","avg_views":{"$avg": "$views"}}}])
list(cur)
```

## Question 6 

```python
cur = collection.aggregate([{"$group":{"_id":"$channel_title","avg_like":{"$avg":"$likes"}}},{"$sort" :{"avg_like": -1}}])
list(cur)
```
